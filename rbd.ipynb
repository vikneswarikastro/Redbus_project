```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/south-bengal-state-transport-corporation-sbstc/?utm_source=rtchometile" )# URL of the website
wait = WebDriverWait(driver, 10)   # Wait for the page to load
driver.maximize_window()
route_links = []
route_names=[]

# Function to scrape bus routes
def scrape_page():
    # Locate elements  (container)
    title=driver.find_elements(By.XPATH,"//a[@class='route']")

    # Loop through each route to extract details
    for route in title:
        try:
            route_names.append(route.text)
            route_links.append(route.get_attribute('href'))

        except Exception as e:
            print(f"An error occurred: {e}")
            continue
```

```python
driver = webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/south-bengal-state-transport-corporation-sbstc/?utm_source=rtchometile")
wait = WebDriverWait(driver, 10)
driver.maximize_window()
for page_number in range(1, 6):
    
    scrape_page()
    if page_number < 5:  # Don't try to click next on the last page
        try:
            # Locate the pagination container
            pagination_container = wait.until(EC.presence_of_element_located(
                (By.XPATH, '//*[@id="root"]/div/div[4]/div[12]') #//*[@id="root"]/div/div[4]/div[12]
            ))

            # Locate the next page button within the container
            next_page_button = pagination_container.find_element(
                By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]'
            )
             # Ensure the next page button is in view
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(1)  # Wait for a bit after scrolling

            # Log the action
            print(f"Clicking on page {page_number + 1}")

            # Click the next page button
            next_page_button.click()

            # Wait for the page number to update to the next page
            wait.until(EC.text_to_be_present_in_element(
                (By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)))

            # Log the successful page navigation
            print(f"Successfully navigated to page {page_number + 1}")

            # Wait for a short duration to ensure the next page loads completely
            time.sleep(3)
        except Exception as e:
            print(f"An error occurred while navigating to page {page_number + 1}: {e}")
            break

```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException   # Existing imports
import time

kd_bus_data = []
driver = webdriver.Chrome()

for index in kd_df.index:
    route = kd_df.loc[index].values[0]
    url = kd_df.loc[index].values[1]
    
    try:
        print(f"Opening URL: {url}")
        driver.get(url)
        wait = WebDriverWait(driver, 40)
        driver.maximize_window()
        time.sleep(5)
    except Exception as e:
        print(f"Error opening URL {url}: {e}")
        continue

    # Handle button clicks
    try:
        b = driver.find_elements(By.CSS_SELECTOR, "div[class='button']")
        print(f"Found {len(b)} buttons")
        for i in range(len(b)-1, -1, -1):
            print(f"Clicking button {i}")
            b[i].click()
    except Exception as e:
        print(f"An error occurred while clicking buttons: {e}")
    
    # Scroll the page
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(10)
        print("Scrolling the page")
    except Exception as e:
        print(f"Error while scrolling: {e}")
    
    try:
        # Wait for bus items to load
        print("Waiting for bus items...")
        bus_item = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "bus-item")))
        time.sleep(5)
        print(f"Found {len(bus_item)} bus items")

        #scrape busdetails
        for bus in bus_item:
            try:
                bus_name = bus.find_element(By.CSS_SELECTOR, "div[class='travels lh-24 f-bold d-color']").text
                bus_type = bus.find_element(By.CSS_SELECTOR, "div[class='bus-type f-12 m-top-16 l-color evBus']").text
                departing_time = bus.find_element(By.CSS_SELECTOR, "div[class='dp-time f-19 d-color f-bold']").text
                duration = bus.find_element(By.CSS_SELECTOR, "div[class='dur l-color lh-24']").text
                reaching_time = bus.find_element(By.CSS_SELECTOR, "div[class='bp-time f-19 d-color disp-Inline']").text
                price = bus.find_element(By.XPATH,"//div[@class='fare d-block']").text
                
                try:
                    star_rating=bus.find_element(By.XPATH, "//div[@class='rating-sec lh-24']").text
                except NoSuchElementException:
                    star_rating= "N/A"
               

                # Check for seat availability with a try-except block
                try:
                    seat_availability = bus.find_element(By.XPATH, "//div[@class='seat-left m-top-16']").text
                except NoSuchElementException:
                    seat_availability = "N/A" # Use a placeholder if seat availability is not found

                # Append data to list
                kd_bus_data.append({
                    "route_name":route,
                    "route_link":url,
                    "bus_name": bus_name,
                    "bus_type": bus_type,
                    "departing_time": departing_time,
                    "duration": duration,
                    "reaching_time": reaching_time,
                    "price": price,
                    "star_rating": star_rating,
                    "seat_availability": seat_availability
                })
                print(f"Scraped data for bus: {bus_name}")
            except Exception as e:
                print(f"An error occurred while scraping bus data: {e}")
    except Exception as e:
        print(f"An error occurred while waiting for bus items: {e}")

print(kd_bus_data)
driver.quit()
```
